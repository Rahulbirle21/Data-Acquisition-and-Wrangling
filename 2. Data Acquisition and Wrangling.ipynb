# Identifying and treating the outliers

```python
Q1 = df4['temp'].quantile(0.25)
Q3 = df4['temp'].quantile(0.75)
```

```python
# IDENTIFYING OUTLIERS INDEXES IN COLUMN 'temp'

def find_outliers(df,cols):
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1
    min_r = Q1 - 1.5 * IQR
    max_r = Q3 + 1.5 * IQR
    outlier_indices = df.index[(df[cols]< min_r) | (df[cols] > max_r)]
    return outlier_indices
    
    
```

```python
find_outliers(df4,'temp')
```

```python
# TRIMMING THE OUTLIERS
df5 = df4[(df4['temp'] >= min_r ) & (df4['temp'] <= max_r)]
df5
```

```python
# Saving the dataset in the directory
df5.to_csv('processed_data.csv')
```
